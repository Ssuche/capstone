```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")

path = '../data/'
```

```python
# import the redcap data
# data\Deidentified REDCap Survey Data 2024.08.26.csv
redcap = pd.read_csv(path + 'Deidentified REDCap Survey Data 2024.08.26.csv')
redcap['visit_date'] = pd.to_datetime(redcap['visit_date']).dt.strftime('%Y-%m-%d')
temp = redcap.copy()
temp['visit_date'] = pd.to_datetime(temp['visit_date'])
temp = temp[['encrypt_mrn', 'ed_screened', 'visit_date', 'new_diagnosis', 'pcp_followup']]
temp = temp.drop_duplicates(subset='encrypt_mrn', keep='first')
```

```python
# import census data
# data\Diabetes Screening Deidentified Geocodes 2024.08.26.csv
census = pd.read_csv(path + 'Diabetes Screening Deidentified Geocodes 2024.08.26.csv', dtype={'zipcode': str})
```

```python
census = census.drop_duplicates(subset='encrypt_mrn', keep='first')

# Perform the left merge
merged = pd.merge(temp, census, on=['encrypt_mrn'], how='left')
```

```python
# import census geocode data
# data\Census Geographic Data File 2023.12.08.csv
geocode = pd.read_csv(path + 'Census Geographic Data File 2023.12.08.csv', dtype={'zipcode': str})
```

```python
merged = pd.merge(merged, geocode, on=['gisjoin'], how='left')
merged = merged.drop_duplicates(subset='encrypt_mrn', keep='first')

```
